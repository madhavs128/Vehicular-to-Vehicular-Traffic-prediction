```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('train_new.csv')
data.tail()
```

```python
data.dropna(inplace = True)
```

```python
cols = ['CellName']
for col in cols:
    data[col] = data[col].map(lambda x: str(x).lstrip('Cell_').rstrip('Cell_')).astype(float)
```

```python
data.head()
```

```python
data = np.array(data)
```

```python
traffic = data[:,-1]
cellname = data[:,-2]
hour = data[:,-3]
year = data[:,-4]
day = data[:,-5]
month = data[:,-6]
```

```python
print(traffic.shape,hour.shape,year.shape,day.shape,month.shape)
np.transpose(traffic)
np.transpose(hour)
np.transpose(year)
np.transpose(day)
np.transpose(month)
print(traffic.shape,hour.shape,year.shape,day.shape,month.shape)
```

```python
tr = []
hr = []
yr = []
da = []
mo = []
for i in range(497542):
    for j in range(1,497542):
        if year[i]==year[j] and month[i]==month[j] and day[i]==day[j] and hour[i] == hour[j]:
            t = traffic[i]+traffic[j]
            tr.append(t)
            hr.append(hour[i])
            yr.append(year[i])
            da.append(day[i])
            mo.append(month[i])
        
    tr.append(traffic[i])
    hr.append(hour[i])
    yr.append(year[i])
    da.append(day[i])
    mo.append(month[i])
```
